<h2> Завдання 2 - 3 ( Варіант № 8 )  </h2>

```python
import numpy as np
import timeit
x = range(3)
x = np.random.standard_normal(3)
def pure_sum():
    return sum(x)
def numpy_sum():
    return np.sum(x)
n = 8
t1 = timeit.timeit(pure_sum, number = n)
print ('Pure Python Sum:', t1)
t2 = timeit.timeit(numpy_sum, number = n)
print ('Numpy Sum:', t2)
```
